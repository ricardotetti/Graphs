```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
import seaborn as sns

#from utils import decorate, savefig

# I set the random seed so the notebook 
# produces the same results every time.
np.random.seed(17)

# TODO: remove this when NetworkX is fixed
from warnings import simplefilter
import matplotlib.cbook
simplefilter("ignore", matplotlib.cbook.mplDeprecation)
from empiricaldist import Pmf
from empiricaldist import Cdf
```

```python
def read_graph(filename):
    G = nx.Graph()
    array = np.loadtxt(filename, dtype=int)
    G.add_edges_from(array)
    return G
fb = read_graph('facebook_combined.txt')
```

```python
degrees = fb.degree()
degree_ = sorted(set(degrees))
degree_value = [lis[1] for lis in degree_] #degree value of each node

#histogram = [(degree_value[i]/float(nx.number_of_nodes(fb)) for i in degree_value]

histogram = []
for i in range(len(degree_value)):
    k = degree_value.count(i)/float(nx.number_of_nodes(fb))
    histogram.append(k)
```

```python
plt.plot(degree_value,histogram,'.')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Degree')
plt.ylabel('Fraction of nodes')
plt.show()
```
